# train test data splitting


```python
 """
    test_ratio:-is the percentage of how much data to be splitted into raining and 
       testing.if test_ratio=0.2 then 20 percent of data is splitted into
       testing remaining 80% to training.
    """
```

```python
def train_test_data_split(data,test_ratio):
  np.random.seed(42)#42 denotes the below explaination.(random state)
#every time we use random we get every time one new shuffled pattern
#so by every time shuffle the test data may mix into train data and turns
 #-the model into overfit
#to avoid that we use random.generator.
#in this case that is np.random.seed(42)
#by this how many times w erun the fuction the shuffed patten of data set will remains same 
  shuffled_Data=np.random.permutation(len(data))
  test_set_size=int(len(data)*test_ratio)
  test_indices=shuffled_Data[:test_set_size]
  training_indices=shuffled_Data[test_set_size:]
  return data.iloc[test_indices] ,data.iloc[training_indices]
```

```python
test_set,train_set=train_test_data_split(data,0.2)
train_set.head()
```

```python
#e=np.random.random(506)#random 506 values in array
#np.arrange(x,y,m)-gives array with values between x and y with a step size of m 
```

```python
"""
!!!https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html

when we split the data,data may not get splitted 
into traing data and testion data uniformly.
for example if  data,if one feature have value counts like
100 people value 0 and 500 people have 1 and if all the the 100 people zero values 
goes into testing data and not even one into training data
then ml model cannot understand that new value.
"""
from sklearn.model_selection import StratifiedShuffleSplit
splitData=StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for training_indices,test_indices in splitData.split(data,data['CHAS']):
  training_set=data.iloc[training_indices]
  test_set=data.iloc[test_indices]
```

# pipeline creating

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
mypip=Pipeline([('imputer',SimpleImputer(strategy='median')),('scalar',StandardScaler())])
data_main=mypip.fit_transform(data,data_labels)
#--pip.fir(data(feature),labels)
data_main.shape
```

# Predicting a desired model:-

```python

from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(data_main,data_labels)

rou_data=data[:5]
rou_labels=data_labels[:5]
rou_ready=mypip.fit_transform (rou_data)
model.predict(rou_ready)

    
```
